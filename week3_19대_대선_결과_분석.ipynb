- 이를 제거하고 **(** 를 기준으로 왼쪽 숫자만 얻어서 콤마(,)를 제거하고, float형으로 변경하는 **get_num()** 함수를 만들겠다

- ```re.split(pattern, string)``` : string을 pattern을 기준으로 나눈다

````python
# re.split() 함수에 대한 이해를 돕기 위한 예시!
import re

test = "수원시에요"
re.split('시', test)
````

````python
import re

def get_num(tmp):
    return float(re.split('\(', str(tmp))[0].replace(',', ''))
````

**아래 모듈을 import해서 ```wait.until``` 함수를 사용할 수도 있다**

````python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

wait = WebDriverWait(driver, 10)

def move_sido(name):
    element = driver.find_element('id',"cityCode")
    element.send_keys(name)
    make_xpath = """//*[@id="searchBtn"]"""
    wait.until(EC.element_to_be_clickable((By.XPATH, make_xpath)))
    driver.find_element('xpath',make_xpath).click()
````

- 미리 준비한 DataFrame에 ```append()``` 명령으로 읽은 데이터를 하나씩 추가해주는 **append_data()** 함수를 만들겠다

image.png

- 이 함수가 실행되면 전체 투표수, 문재인후보, 홍준표후보, 안철수후보의 득표수가 추가된다

````python
def append_data(df, sido_name, data):
    for each in df[0].values[1:]:
        data['광역시도'].append(sido_name)
        data['시군'].append(each[0])
        data['pop'].append(get_num(each[2]))
        data['moon'].append(get_num(each[3]))
        data['hong'].append(get_num(each[4]))
        data['ahn'].append(get_num(each[5]))
````

## 각 후보의 득표율과 지역 ID 정리하기

````python
election_result = pd.read_csv("./data/election_result.csv", encoding = 'utf-8', index_col = 0)
election_result.head()
````

- '광역시도' 이름에서 앞 두 글자만 따오겠다

````python
sido_candi = election_result['광역시도']
sido_candi = [name[:2] if name[:2] in ['서울', '부산', '대구', '광주', '인천', '대전', '울산']
                          else '' for name in sido_candi]
````

- '중구', '남구'와 같은 두 글자 이름은 그대로 보내고, '중랑구', '서초구'와 같은 두 글자 이상의 이름은 '중랑', '서초'와 같이 줄이기 위한 **cut_char_sigu()** 함수를 만들겠다

````python
def cut_char_sigu(name):
    return name if len(name) == 2 else name[:-1]
````

**광역시가 아닌데 행정구를 가지고 있는 '시군'에 대해 처리해주겠다**
- **'수원', '성남', '안양', '안산', '고양', '용인', '청주', '천안', '전주', '포항', '창원'**에 대해 **'안양 만안', '안양 동안'**과 같이 정리!!

````python
import re

sigun_candi = [''] * len(election_result)

for n in election_result.index:
    each = election_result['시군'][n]
    if each[:2] in ['수원', '성남', '안양', '안산', '고양', 
                    '용인', '청주', '천안', '전주', '포항', '창원']:
        sigun_candi[n] = re.split('시', each)[0] + ' ' + \
                         cut_char_sigu(re.split('시', each)[1])
    else:
        sigun_candi[n] = cut_char_sigu(each)

sigun_candi
````

- sido_candi 변수에서 그냥 공란이 있으면 첫 글자가 띄어쓰기가 될 수 있다
    - 따라서 첫 글자가 공백인 경우, 그 다음 글자부터 읽어오도록 해주었다
- 또한, **'세종시'**의 경우는 예외로 따로 처리해주었다

````python
ID_candi = [sido_candi[n] + ' ' + sigun_candi[n] for n in range(0, len(sigun_candi))]

ID_candi = [name[1:] if name[0] == ' ' else name for name in ID_candi]
ID_candi = [name[:2] if name[:2] == '세종' else name for name in ID_candi]

ID_candi
````

**최종 데이터 셋 생성!!**
- **election_result**와 **draw_korea**를 ```merge()```를 사용해서 최종 데이터 셋을 생성하겠다
    - election_result에 draw_korea를 **'ID'를 기준**으로 **left join** 시키겠다
    - pd.merge() : 데이터프레임 병합

````python
final_elect_data = pd.merge(election_result, draw_korea, how = 'outer', on = ['ID'] )
final_elect_data.head()
````

**각 후보들 간의 득표율 차이를 계산**

## 19대 대선 결과 득표율 시각화하기
- 먼저 지도의 경계선을 그려주기 위해, BORDER_LINE 변수를 생성해주겠다
- 지도를 그릴 ```drawKorea()``` 함수를 만들어 줄 것이다

````python
BORDER_LINES = [
    [(5, 1), (5,2), (7,2), (7,3), (11,3), (11,0)], # 인천
    
    [(5,4), (5,5), (2,5), (2,7), (4,7), (4,9), (7,9), 
     (7,7), (9,7), (9,5), (10,5), (10,4), (5,4)], # 서울
    
    [(1,7), (1,8), (3,8), (3,10), (10,10), (10,7), 
     (12,7), (12,6), (11,6), (11,5), (12, 5), (12,4), 
     (11,4), (11,3)], # 경기도
    
    [(8,10), (8,11), (6,11), (6,12)], # 강원도
    
    [(12,5), (13,5), (13,4), (14,4), (14,5), (15,5), 
     (15,4), (16,4), (16,2)], # 충청북도
    
    [(16,4), (17,4), (17,5), (16,5), (16,6), (19,6), 
     (19,5), (20,5), (20,4), (21,4), (21,3), (19,3), (19,1)], # 전라북도
    
    [(13,5), (13,6), (16,6)], # 대전시
    
    [(13,5), (14,5)], # 세종시
    
    [(21,2), (21,3), (22,3), (22,4), (24,4), (24,2), (21,2)], # 광주
    
    [(20,5), (21,5), (21,6), (23,6)], # 전라남도
    
    [(10,8), (12,8), (12,9), (14,9), (14,8), (16,8), (16,6)], # 충청북도
    
    [(14,9), (14,11), (14,12), (13,12), (13,13)], # 경상북도
    
    [(15,8), (17,8), (17,10), (16,10), (16,11), (14,11)], # 대구
    
    [(17,9), (18,9), (18,8), (19,8), (19,9), (20,9), (20,10), (21,10)], # 부산
    
    [(16,11), (16,13)], # 울산
    
    [(27,5), (27,6), (25,6)],
]
````

````python
def drawKorea(targetData, blockedMap, cmapname):
    gamma = 0.75

    whitelabelmin = 20.

    datalabel = targetData

    tmp_max = max([np.abs(min(blockedMap[targetData])), np.abs(max(blockedMap[targetData]))])
    vmin, vmax = -tmp_max, tmp_max

    mapdata = blockedMap.pivot_table(index = 'y', columns = 'x', values = targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize = (9, 11))
    plt.pcolor(masked_mapdata, vmin = vmin, vmax = vmax, cmap = cmapname, 
               edgecolor = '#aaaaaa', linewidth = 0.5)

    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():
        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다
        # (중구, 서구)
        if len(row['ID'].split()) == 2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2] == '고성':
            dispname = '고성'
        else:
            dispname = row['ID']

        # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11, 1.

        annocolor = 'white' if np.abs(row[targetData]) > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x'] + 0.5, row['y'] + 0.5), weight = 'bold',
                     fontsize = fontsize, ha = 'center', va = 'center', color = annocolor,
                     linespacing = linespacing)

    # 시도 경계 그린다
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c = 'black', lw = 2)

    plt.gca().invert_yaxis()

    plt.axis('off')

    cb = plt.colorbar(shrink = 0.1, aspect = 10)
    cb.set_label(datalabel)

    plt.tight_layout()
    plt.show()
````

**"문재인 후보 vs 홍준표 후보" 득표율 격차를 시각화**
- **문재인 후보**의 경우, **전라도**의 압도적인 지지율을 확인할 수 있다. 또한 **서울, 경기권**의 경우에도 문재인 후보의 지지율이 높은 것을 알 수 있다
- 반면, **홍준표 후보**는 **대구, 경북** 지역에서 우세를 보인다
- 하지만 **경남 창원, 김해, 부산** 지역은 **문재인 후보**가 홍준표 후보보다 **지지율이 높다**

````python
drawKorea('moon_vs_hong', final_elect_data, 'RdBu')
````

**"문재인 후보 vs 안철수 후보" 득표율 격차를 시각화**
- 이건 뭐...딱 봐도 **문재인 후보**의 **대승**이다!!

````python
drawKorea('moon_vs_ahn', final_elect_data, 'RdBu')
````

**"안철수 후보 vs 홍준표 후보" 득표율 격차를 시각화**
- 안철수 후보와 홍준표 후보의 지지율은 거의 **박빙**이다
- **안철수 후보**는 **전라도** 지역의 지지율이 높고, **홍준표 후보**는 **대구, 경북** 지역의 지지율이 높다

````python
drawKorea('ahn_vs_hong', final_elect_data, 'RdBu')
````
