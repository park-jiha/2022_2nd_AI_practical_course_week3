```python
# iloc
# 데이터 프레임의 행이나 칼럼의 순서를 나타내는 정수로 특정 값을 추출해오는 방법
# 기본 형식 = df.iloc[행 index, 열 index]
print(df.iloc[0:4])
```

```python
print(df.iloc[:,0:4])
```

```python
# iterrows 를 이용해 원하는 정보 찾기
# iterrows 객체는 데이터프레임의 각 행의 정보를 담은 객체이다.
print(df.iterrows)
for index, row in df.iterrows():
  print(index, row['Name'])
```

```python
# loc
# 데이터프레임의 행이나 칼럼의 라벨을 boolean array로 인덱싱하는 방법
# 즉, 우리가 읽을 수 있는 라벨의 값으로 특정 값들을 골라오는 방법
# 기본 형식 = df.loc[행 index 값, 열 index 값]
df.loc[df['Type 1'] == "Grass"]
```

### 데이터 수정

```python
df.head(5)
```

```python
# 새로운 항목 만들기
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df
```

```python
# column 삭제 (drop)
df = df.drop(columns = ['Total'])
df
```

```python
# 항목을 언제 다 적어! -> iloc 활용
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df
```

```python
# 원하는 곳에 배치시키고 싶을 때
cols = list(df.columns)
print(cols)

df = df[cols[0:4] + [cols[-1]] + cols[4:-1]]
df
```

### 데이터 저장

```python
# csv 형태로 저장
df.to_csv('./data/modified.csv', index=False)
```

```python
# xlsx 형태로 저장
df.to_excel('./data/modified.xlsx', index=False)
```

```python
# txt 형태로 저장('\t' 구분 주의)
df.to_csv('./data/modified.txt', index=False, sep='\t')
```

### 데이터 필터링

```python
df
```

```python
# loc을 활용해 조건에 맞는 index만 추출해보자
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
new_df
```

```python
# index 재정비
# drop : 새롭게 추가되는 index를 삭제할지, 말지
# inplace : 원본 객체를 변경할지, 말지
new_df.reset_index(drop=True, inplace=True)
new_df.to_csv('./data/filtered.csv')
new_df
```

### 세부 항목 수정

```python
# loc을 활용하여 세부 항목 수정
df.loc[df['Total'] > 500, ['Generation','Legendary']] = ['Test 1', 'Test 2']
df
```

```python
# Grouping을 통한 데이터 분석(갯수 세어보기 count)
df = pd.read_csv('./data/modified.csv')

df.groupby(['Type 1']).count()['Name']
#df.groupby(['Type 1', 'Type 2']).count()['Name']
```
